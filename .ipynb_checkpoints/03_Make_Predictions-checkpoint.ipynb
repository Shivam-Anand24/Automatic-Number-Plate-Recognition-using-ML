```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import load_img, img_to_array
from tensorflow.keras.models import load_model

def object_detection(input_path, output_path):
    # Load the pre-trained object detection model
    model_path = "C:/Users/shiva/OneDrive/Desktop/anpr/models/object_detection.h5"
    model = load_model(model_path)

    # Get a list of image files in the input directory
    image_files = os.listdir(input_path)
    for file_name in image_files:
        if file_name.endswith('.jpg'):
            # Construct the full path to the input image
            image_path = os.path.join(input_path, file_name)
            
            # Load the image using OpenCV
            image = cv2.imread(image_path)
            
            # Load the image with target size for the model
            image1 = load_img(image_path, target_size=(224, 224))
            image_arr_224 = img_to_array(image1) / 255.0
            
            # Get the original dimensions of the image
            h, w, d = image.shape
            
            # Reshape and preprocess the image for model prediction
            test_arr = image_arr_224.reshape(1, 224, 224, 3)
            
            # Predict object coordinates using the model
            coords = model.predict(test_arr)
            
            # Denormalize the predicted coordinates
            denorm = np.array([w, w, h, h])
            coords = coords * denorm
            coords = coords.astype(np.int32)
            xmin, xmax, ymin, ymax = coords[0]
            
            # Define the top-left and bottom-right points of the rectangle
            pt1 = (xmin, ymin)
            pt2 = (xmax, ymax)
            
            # Draw a rectangle on the image
            cv2.rectangle(image, pt1, pt2, (0, 255, 0), 3)
            
            # Construct the output image path
            output_image_path = os.path.join(output_path, file_name)
            
            # Save the image with the drawn rectangle
            cv2.imwrite(output_image_path, image)

# Input and output folders
input_folder = 'C:/Users/shiva/OneDrive/Desktop/anpr/dataset'
output_folder = 'C:/Users/shiva/OneDrive/Desktop/anpr/output'

# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)

# Call the object_detection function with input and output folders
object_detection(input_folder, output_folder)
```
